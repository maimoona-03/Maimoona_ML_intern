## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

My lane is a clustering task. This means I'm not predicting a known answer — I'm letting the computer find natural groups of similar pages on its own, based on their numbers.


## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

There is no single "target" number to predict here. Instead, the output is the cluster group each page gets put into (like Group 1, Group 2, Group 3, Group 4). The computer creates these groups itself — they're not already labeled in the data.


## 3. Success metric

*One metric you can defend. What number means 'good'?*

Since there's no "correct answer" to check against, I'll judge success two ways:

A number called the silhouette score, which tells me how neatly separated the groups are.
A simple gut check: do the groups actually make sense? For example, does one group clearly look like "popular pages" and another like "old, ignored pages"?

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

I could try writing a simple rule, like "if a page has low traffic AND is old, call it weak." But with hundreds of thousands of pages and multiple numbers changing together, a fixed rule like this is too simple and misses a lot of patterns. Clustering lets the computer look at all the numbers together and find groups that a person would never spot just by eyeballing spreadsheets.
